```python
from google.colab import files
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import RandomForestClassifier
from sklearn import metrics
```

```python
df = pd.read_csv("mp data.csv.csv")
```

```python
df.head()
```

```python
inputs = df.drop('result', axis='columns')
target = df['result']
```

```python
inputs.head()
```

```python
target.head()
```

```python
le_gender = LabelEncoder()
le_underlying_disease = LabelEncoder()
le_infection = LabelEncoder()
le_medication = LabelEncoder()
le_result = LabelEncoder()  # LabelEncoder for 'result'
```

```python
inputs['gender_n'] = le_gender.fit_transform(inputs['gender'])
inputs['underlying_disease_n'] = le_underlying_disease.fit_transform(inputs['underlying_disease'])
inputs['infection_n'] = le_infection.fit_transform(inputs['infection'])
inputs['medication_n'] = le_medication.fit_transform(inputs['medication'])
target_encoded = le_result.fit_transform(target)  # Encoding 'result'
```

```python
print(inputs)
```

```python
inputs_n = inputs.drop(['serial_number', 'gender', 'underlying_disease', 'infection', 'medication'], axis='columns')
```

```python
print(inputs_n)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target_encoded, test_size=0.2,random_state=42)
```

```python
print(X_test.head())
for i in range(0,5):
  print(y_test[i])
```

```python
model = RandomForestClassifier(max_features=9)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, predictions)
print("Accuracy Score =", accuracy)
```

```python
example_input = [1, 9, 15, 30, 1, 0, 0, 2, 3]
predicted_result = model.predict([example_input])
print(f"Predicted Result: {predicted_result}")
```
